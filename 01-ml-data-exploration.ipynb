# ML Data Exploration

Ce notebook présente des étapes de base pour explorer un jeu de données de machine learning avec `pandas`, `seaborn` et `matplotlib`.

Il utilise le jeu de données `iris` de `scikit-learn` comme exemple.

```python
# Installation des dépendances recommandées
# pip install pandas seaborn scikit-learn matplotlib
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

sns.set(style="whitegrid")
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

## Aperçu du jeu de données

Affichons les dimensions et des statistiques descriptives du dataset.

```python
print('Dimensions :', df.shape)
print('
Colonnes :', df.columns.tolist())

df.describe()
```

## Répartition des classes

Analyse de la fréquence des classes cibles.

```python
df['species'].value_counts()
```

## Visualisation des relations entre variables

Nous utilisons un pairplot pour observer les relations entre les variables supplémentaires.

```python
sns.pairplot(df, hue="species", vars=iris.feature_names)
plt.suptitle('Pairplot du dataset Iris', y=1.02)
plt.show()
```

## Heatmap de corrélation

Un aperçu des corrélations entre les caractéristiques.

```python
corr = df[iris.feature_names].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=.2f)
plt.title('Matrice de corrélation')
plt.show()
```
